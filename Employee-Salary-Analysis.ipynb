```python
#Importing the required Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset

df= pd.read_csv("Employers_data.csv")
```

```python
#Disp[lay the first few rows
df.head()
```

```python
# count of employees in each department

dept_counts = df['Department'].value_counts()

#Plotting

plt.figure(figsize=(8,6))
sns.barplot(x=dept_counts.index, y=dept_counts.values)
plt.title("Employee Count By Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
## 📊 Department-Wise Employee Distribution

## Understanding how employees are distributed across departments gives insight into the company's workforce structure. This bar chart clearly shows which departments are the most and least staffed.
```

```python
# calculate average salry by education level
edu_salary = df.groupby("Education_Level")["Salary"].mean().sort_values(ascending=False)

#Plotting

plt.figure(figsize=(8,6))
sns.barplot(x=edu_salary.index, y=edu_salary.values)
plt.title("Average Salary by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Average Salary (£)")
plt.tight_layout()
plt.show()
```

```python
# calcuate the average salary by department

dept_salary=df.groupby("Department")["Salary"].mean().sort_values(ascending=False)

#ploting 
plt.figure(figsize=(10,6))
sns.barplot(x=dept_salary.index, y=dept_salary.values)
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary(£)")
plt.tight_layout()
plt.show()
```

```python
## 🏢 Average Salary by Department

# This bar chart reveals the average salary offered across various departments. It helps identify which departments are the most highly compensated and can inform decisions about career focus or budget allocation.
```

```python
# Calculate average salary by gender
gender_salary = df.groupby("Gender")["Salary"].mean()

# Plotting
plt.figure(figsize=(6, 5))
sns.barplot(x=gender_salary.index, y=gender_salary.values)
plt.title("Average Salary by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Salary (£)")
plt.tight_layout()
plt.show()

# Print the values for reference
print("Average Salary by Gender:\n", gender_salary)
```

```python
## 👩‍💼👨‍💼 Gender Pay Gap Analysis

# This chart compares the average salary of male and female employees. Understanding pay differences between genders helps promote transparency and can highlight opportunities for creating a more equitable workplace.
```

```python
# Calculate average salary by department and gender
dept_gender_salary = df.groupby(['Department', 'Gender'])['Salary'].mean().unstack()

# Plotting
dept_gender_salary.plot(kind='bar', figsize=(10, 6))
plt.title("Average Salary by Gender Within Each Department")
plt.xlabel("Department")
plt.ylabel("Average Salary (£)")
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.tight_layout()
plt.show()

# Print the DataFrame
print("Average Salary by Department and Gender:\n")
print(dept_gender_salary)
```

```python
# Select relevant numeric columns
corr_data = df[['Age', 'Experience_Years', 'Salary']]

# Calculate correlation matrix
corr_matrix = corr_data.corr()

# Display the correlation matrix
print("Correlation Matrix:\n")
print(corr_matrix)

# Plot heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation between Age, Experience, and Salary")
plt.tight_layout()
plt.show()
```

```python
## 📈 Correlation Analysis: Age, Experience, and Salary

# This heatmap visualizes the strength of correlation between age, years of experience, and salary. A higher correlation means that an increase in one variable tends to be associated with an increase in another. This helps identify what factors most influence employee compensation.
```

```python
# Calculate average salary by location
location_salary = df.groupby("Location")["Salary"].mean().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=location_salary.index, y=location_salary.values)
plt.title("Average Salary by Location")
plt.xlabel("Location")
plt.ylabel("Average Salary (£)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Optional: Display the actual values
print("Average Salary by Location:\n")
print(location_salary)
```

```python
# Histogram to show salary distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], kde=True, bins=30)
plt.title("Distribution of Salaries")
plt.xlabel("Salary (£)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Boxplot to identify salary outliers
plt.figure(figsize=(10, 2))
sns.boxplot(x=df['Salary'], color='skyblue')
plt.title("Boxplot of Salaries")
plt.xlabel("Salary (£)")
plt.tight_layout()
plt.show()
```

```python
## 💰 Salary Distribution Analysis

 #The histogram helps us understand the distribution of salaries in the dataset — whether it's skewed, normal, or multi-modal. The boxplot complements this by highlighting outliers and spread.

# Key observations:
# - Most salaries fall between £70,000 and £150,000.
# - A few outliers exist on the higher end, possibly indicating senior roles or executive-level positions.
```
